### Solar System planets evolution (no heat from the core) (Fig. 7)

```python
yr = 365.0*24.0*60.0*60.0   # 1 year in seconds
        
fig = plt.figure(figsize=(8.5,15.75))    
small_size = 15
medium_size = 20
bigger_size = 20

lw = 2.5
    
plt.rc('font', size=small_size)          # controls default text sizes
plt.rc('axes', titlesize=small_size)     # fontsize of the axes title
plt.rc('axes', labelsize=medium_size)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=medium_size)   # fontsize of the tick labels
plt.rc('ytick', labelsize=medium_size)   # fontsize of the tick labels
plt.rc('legend', fontsize=medium_size)   # legend fontsize
plt.rc('figure', titlesize=bigger_size)  # fontsize of the figure title

nx_panels = 3
ny_panels = 1
n = 0

labelx = 1
labely = 1.03

ymin = 1300
ymax = 2150

############################################
n = n + 1
ax = fig.add_subplot(nx_panels,ny_panels,n)
ax.plot(q_venus_e19.t[:-1]/yr/1e6, T_abovesol_venus_e19, '-', color=venus_col, label='Venus (SL)',lw=lw)
ax.plot(q_venus_e19.t[:-1]/yr/1e6, T_belowsol_venus_e19, '--', color=venus_col, label='',lw=lw)
ax.plot(q_mars_e19.t[:-1]/yr/1e6, T_abovesol_mars_e19, '-', color=mars_col, label='Mars (SL)',lw=lw)
ax.plot(q_mars_e19.t[:-1]/yr/1e6, T_belowsol_mars_e19, '--', color=mars_col, label='',lw=lw)
ax.plot(q_mercury_e19.t[:-1]/yr/1e6, T_abovesol_mercury_e19, '-', color=mercury_col, label='Mercury (SL)',lw=lw)
ax.plot(q_mercury_e19.t[:-1]/yr/1e6, T_belowsol_mercury_e19, '--', color=mercury_col, label='',lw=lw)
ax.plot(q_earth_e19.t[:-1]/yr/1e6, T_abovesol_earth_e19, '-', color=earth_col, label='Earth (ML)',lw=lw)
ax.plot(q_earth_e19.t[:-1]/yr/1e6, T_belowsol_earth_e19, '--', color=earth_col, label='',lw=lw)
# ax.plot(q_moon_e19.t[:-1]/yr/1e6, T_abovesol_moon_e19, '-', color=moon_col, label='Moon',lw=lw)
# ax.plot(q_moon_e19.t[:-1]/yr/1e6, T_belowsol_moon_e19, '--', color=moon_col, label='',lw=lw)

ax.set_ylim(ymin,ymax)
ax.grid()
ax.set_ylabel('Mantle temperature [K]')
ax.set_xlabel('Time [Myr]')
ax.set_title('$\mu_{0}=10^{19}$ Pa s', fontsize=bigger_size)
ax.legend(loc=1)

ax.text(labelx, labely, 'a', ha='right', va='center', transform=ax.transAxes, fontsize=bigger_size)

############################################
n = n + 1
ax = fig.add_subplot(nx_panels,ny_panels,n)
ax.plot(q_mercury_e20.t[:-1]/yr/1e6, T_abovesol_mercury_e20, '-', color=mercury_col, label='Mercury',lw=lw)
ax.plot(q_mercury_e20.t[:-1]/yr/1e6, T_belowsol_mercury_e20, '--', color=mercury_col, label='',lw=lw)
ax.plot(q_venus_e20.t[:-1]/yr/1e6, T_abovesol_venus_e20, '-', color=venus_col, label='Venus',lw=lw)
ax.plot(q_venus_e20.t[:-1]/yr/1e6, T_belowsol_venus_e20, '--', color=venus_col, label='',lw=lw)
ax.plot(q_earth_e20.t[:-1]/yr/1e6, T_abovesol_earth_e20, '-', color=earth_col, label='Earth',lw=lw)
ax.plot(q_earth_e20.t[:-1]/yr/1e6, T_belowsol_earth_e20, '--', color=earth_col, label='',lw=lw)
# ax.plot(q_moon_e20.t[:-1]/yr/1e6, T_abovesol_moon_e20, '-', color=moon_col, label='Moon',lw=lw)
# ax.plot(q_moon_e20.t[:-1]/yr/1e6, T_belowsol_moon_e20, '--', color=moon_col, label='',lw=lw)
ax.plot(q_mars_e20.t[:-1]/yr/1e6, T_abovesol_mars_e20, '-', color=mars_col, label='Mars',lw=lw)
ax.plot(q_mars_e20.t[:-1]/yr/1e6, T_belowsol_mars_e20, '--', color=mars_col, label='',lw=lw)

ax.set_ylim(ymin,ymax)
ax.grid()
ax.set_ylabel('Mantle temperature [K]')
ax.set_xlabel('Time [Myr]')
ax.set_title('$\mu_{0}=10^{20}$ Pa s', fontsize=bigger_size)

ax.text(labelx, labely, 'b', ha='right', va='center', transform=ax.transAxes, fontsize=bigger_size)

############################################
n = n + 1
ax = fig.add_subplot(nx_panels,ny_panels,n)
ax.plot(q_mercury_e21.t[:-1]/yr/1e6, T_abovesol_mercury_e21, '-', color=mercury_col, label='Mercury',lw=lw)
ax.plot(q_mercury_e21.t[:-1]/yr/1e6, T_belowsol_mercury_e21, '--', color=mercury_col, label='',lw=lw)
ax.plot(q_venus_e21.t[:-1]/yr/1e6, T_abovesol_venus_e21, '-', color=venus_col, label='Venus',lw=lw)
ax.plot(q_venus_e21.t[:-1]/yr/1e6, T_belowsol_venus_e21, '--', color=venus_col, label='',lw=lw)
ax.plot(q_earth_e21.t[:-1]/yr/1e6, T_abovesol_earth_e21, '-', color=earth_col, label='Earth',lw=lw)
ax.plot(q_earth_e21.t[:-1]/yr/1e6, T_belowsol_earth_e21, '--', color=earth_col, label='',lw=lw)
# ax.plot(q_moon_e21.t[:-1]/yr/1e6, T_abovesol_moon_e21, '-', color=moon_col, label='Moon',lw=lw)
# ax.plot(q_moon_e21.t[:-1]/yr/1e6, T_belowsol_moon_e21, '--', color=moon_col, label='',lw=lw)
ax.plot(q_mars_e21.t[:-1]/yr/1e6, T_abovesol_mars_e21, '-', color=mars_col, label='Mars',lw=lw)
ax.plot(q_mars_e21.t[:-1]/yr/1e6, T_belowsol_mars_e21, '--', color=mars_col, label='',lw=lw)

ax.set_ylim(ymin,ymax)
ax.grid()
ax.set_ylabel('Mantle temperature [K]')
ax.set_xlabel('Time [Myr]')
ax.set_title('$\mu_{0}=10^{21}$ Pa s', fontsize=bigger_size)

ax.text(labelx, labely, 'c', ha='right', va='center', transform=ax.transAxes, fontsize=bigger_size)

plt.tight_layout()
plt.savefig('../Examples/Figures/Fig7.pdf', format='pdf')
```
