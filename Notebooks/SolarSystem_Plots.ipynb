```python
Ts = 300.

Tm = 1700.
Tm_min = 1500.
Tm_max = 1900.

dT = Tm - Ts
dT_min = Tm_min - Ts
dT_max = Tm_max - Ts

rho = 4000.
rho_min = 3000.
rho_max = 5000. 

g = 10.
g_min = 2.
g_max = 15.

C = 0.5
k = 3.
alpha = 3e-5
kappa = 1e-6
Racr = 1100
E = 3e5
R = 8.314

theta = E*dT/R/Tm**2
theta_min = E*dT/R/Tm_max**2
theta_max = E*dT/R/Tm_min**2

eta_arr = np.logspace(18,22,20)

qs_ML = C*k*dT*(rho*g*alpha*dT/eta_arr/kappa/Racr)**(1./3)
qs_SL = C*k*dT*theta**(-4./3)*(rho*g*alpha*dT/eta_arr/kappa/Racr)**(1./3)

qs_ML_min = C*k*dT_min*(rho_min*g_min*alpha*dT_min/eta_arr/kappa/Racr)**(1./3)
qs_SL_min = C*k*dT_min*theta_max**(-4./3)*(rho_min*g_min*alpha*dT_min/eta_arr/kappa/Racr)**(1./3)

qs_ML_max = C*k*dT_max*(rho_max*g_max*alpha*dT_max/eta_arr/kappa/Racr)**(1./3)
qs_SL_max = C*k*dT_max*theta_min**(-4./3)*(rho_max*g_max*alpha*dT_max/eta_arr/kappa/Racr)**(1./3)
```

```python
fig = plt.figure(figsize=(8,5.5))    
small_size = 10
medium_size = 15
bigger_size = 20

lw = 2.5 

mcolor = 'tab:blue'
bcolor = 'tab:orange'
ccolor = 'tab:red'
    
plt.rc('font', size=small_size)          # controls default text sizes
plt.rc('axes', titlesize=small_size)     # fontsize of the axes title
plt.rc('axes', labelsize=medium_size)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=medium_size)   # fontsize of the tick labels
plt.rc('ytick', labelsize=medium_size)   # fontsize of the tick labels
plt.rc('legend', fontsize=medium_size)   # legend fontsize
plt.rc('figure', titlesize=bigger_size)  # fontsize of the figure title

nx_panels = 1
ny_panels = 1
n = 0

############################################
# Scaling laws
############################################
n = n + 1
ax = fig.add_subplot(nx_panels,ny_panels,n)
ax.plot(eta_arr, qs_ML*1e3, color=mcolor, label='Mobile lid', lw=lw)
# ax.plot(eta_arr, qs_ML_min*1e3, '--', color=mcolor, lw=lw)
# ax.plot(eta_arr, qs_ML_max*1e3, '--', color=mcolor, lw=lw)
ax.fill_between(eta_arr, qs_ML_min*1e3, qs_ML_max*1e3, facecolor=mcolor, alpha=0.3)

ax.plot(eta_arr, qs_SL*1e3, color=ccolor, label='Stagnant lid', lw=lw)
# ax.plot(eta_arr, qs_SL_min*1e3, '--', color=ccolor,  lw=lw)
# ax.plot(eta_arr, qs_SL_max*1e3, '--', color=ccolor,  lw=lw)
ax.fill_between(eta_arr, qs_SL_min*1e3, qs_SL_max*1e3, facecolor=ccolor, alpha=0.3)

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('Viscosity [Pa s]')
ax.set_ylabel('Heat flux [mW/m$^2$]')
ax.grid()
ax.legend(loc=1)

plt.tight_layout()
plt.savefig('Figures/heat_fluxes.pdf', format='pdf')
```
