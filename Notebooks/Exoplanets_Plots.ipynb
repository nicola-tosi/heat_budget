## SOLIDUS CHANGES WITH INCREASING MASS

```python
# Maximum pressure for melting
Pmax = 10e9

# Surface gravity for different masses
g_M05 = 7.13
g_M1 = 9.8
g_M2 = 13.49
g_M4 = 18.56

# Mantle density for different masses
rho_M05 = 3929
rho_M1 = 4460
rho_M2 = 5018
rho_M4 = 5676

# Arrays of depth down to Pmax...
z_M05 = np.linspace(0, Pmax/(rho_M05*g_M05),20)
z_M1 = np.linspace(0, Pmax/(rho_M1*g_M1),20)
z_M2 = np.linspace(0, Pmax/(rho_M2*g_M2),20)
z_M4 = np.linspace(0, Pmax/(rho_M4*g_M4),20)

# ...and corresponding pressure increase
P_M05 = rho_M05*g_M05*z_M05
P_M1 = rho_M1*g_M1*z_M1
P_M2 = rho_M2*g_M2*z_M2
P_M4 = rho_M4*g_M4*z_M4

# Solidus down to depth corresponding to Pmax for different masses
Tsol_M05 = suppf.calculate_dry_solidus(P_M05/1e9)
Tsol_M1 = suppf.calculate_dry_solidus(P_M1/1e9)
Tsol_M2 = suppf.calculate_dry_solidus(P_M2/1e9)
Tsol_M4 = suppf.calculate_dry_solidus(P_M4/1e9)
```

```python
# Select one simulation and the corresponding time step
s1 = q_M1_Q24_T1700_e19_SL
i = 300

# Number of grid points
npts = 20 # in the upper boundary layer
npta = 20 # in the mantle

# Pressure at the top and bottom TBLs
Pm1 = s1.rhom*s1.g*s1.delta_s[i]
Pb1 = s1.rhom*s1.g*(s1.Rp - (s1.Rc + s1.delta_c[i]))
# Adiabatic mantle between top and bottom TBLs
ra1 = np.linspace(s1.Rc + s1.delta_c[i], s1.Rp - s1.delta_s[i],  npta)
Ta1 = suppf.calculate_adiabat(s1, npta, s1.Tm[i], Pm1, Pb1)
Ta1 = np.flipud(Ta1)
# Temperature across the top TBL
rs1 = np.linspace(s1.Rp - s1.delta_s[i], s1.Rp, npts)
Trs1 = np.linspace(Ta1[npta-1], s1.Ts, npts)
# Thermal expansivity at the top TBL
alpha1 = suppf.calculate_thermal_expansivity(s1.Tm[i], Pm1/1e9)

# Adiabatic mantle between the surface and the bottom TBL
raa = np.linspace(s1.Rc + s1.delta_c[i], s1.Rp , npta)
# I Added a +20 by hand because the figure was not plotting correctly. Check what's going on!
Tp = s1.Tm[i]*np.exp(-alpha1*s1.g*s1.delta_s[i]/s1.cm) + 20 # potential temperature 
Taa = suppf.calculate_adiabat(s1, npta, Tp, 0., Pb1)
Taa = np.flipud(Taa)
```
