```python
def calculate_initial_concentrations(X_U,X_Th,X_K,t):  
    """Given present-day bulk abundances, calculate initial bulk concentrations"""
    
    yr = 365.0*24.0*60.0*60.0
    t = t*yr
    
    e = lambda tbp, tau : np.exp(np.log(2.)*tbp/tau)
    
    tau_U238  = 4.47e9*yr
    tau_U235  = 0.704e9*yr
    tau_K40   = 1.2483e9*yr
    tau_Th232 = 14.5e9*yr
    
    f_U238  = 0.9928
    f_U235  = 7.2e-3
    f_Th232 = 1.0
    f_K40   = 1.19e-4

    X0_U238  = X_U*f_U238*e(t,tau_U238)
    X0_U235  = X_U*f_U235*e(t,tau_U235)
    X0_Th232 = X_Th*f_Th232*e(t,tau_Th232)    
    X0_K40   = X_K*f_K40*e(t,tau_K40)
        
#     X0_U238  = X_U*f_U238*np.exp(np.log(2.)*t/tau_U238)
#     X0_U235  = X_U*f_U235*np.exp(np.log(2.)*t/tau_U235)
#     X0_Th232 = X_Th*f_Th232*np.exp(np.log(2.)*t/tau_Th232)    
#     X0_K40   = X_K*f_K40*np.exp(np.log(2.)*t/tau_K40)    

    return X0_U238, X0_U235, X0_Th232, X0_K40
```
